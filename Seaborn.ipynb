## Seaborn

* Powerful data visualization library
* Closely integrated with pandas
* it builts on top of matplotlib


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style='darkgreen')
```

```python
data=sns.load_dataset('tips') #seaborn inbuilt datasets : https://github.com/mwaskom/seaborn-data
print(data.head())
print(type(data))
```

```python
sns.relplot(x='total_bill',y='tip',data=data) #relplot--->relational plot between two varible
```

* there is linear relation between total_bill and tip
* as the total_bill increase there is chance for getting more tip

```python
sns.relplot(x="total_bill",y="tip",data=data,hue='smoker')
```

```python
#avg totalbill of smoker==Yes
print(sum(data[data['smoker']=='Yes'] ['total_bill'])/len(data[data['smoker']=='Yes'] ['total_bill']))
# avg totalbills of smoker== No
print(sum(data[data['smoker']=='No']['total_bill'])/len(data[data['smoker']=='No'] ['total_bill']))

#avg tip of smoker==yes
print(sum(data[data['smoker']=='Yes'] ['tip'])/len(data[data['smoker']=='Yes'] ['tip']))
#avg tip of smoker==No
print(sum(data[data['smoker']=='No']['tip'])/len(data[data['smoker']=='No'] ['tip']))
```

```python
sns.relplot(x="total_bill",y="tip",data=data,hue='sex')
```

```python
#avg totalbill of sex==male
print(sum(data[data['sex']=='Male'] ['total_bill'])/len(data[data['sex']=='Male'] ['total_bill']))
# avg totalbills of smoker== No
print(sum(data[data['sex']=='Female']['total_bill'])/len(data[data['sex']=='Female'] ['total_bill']))

#avg tip of smoker==yes
print(sum(data[data['sex']=='Male'] ['tip'])/len(data[data['sex']=='Male'] ['tip']))
#avg tip of smoker==No
print(sum(data[data['sex']=='Female']['tip'])/len(data[data['sex']=='Female'] ['tip']))
```
